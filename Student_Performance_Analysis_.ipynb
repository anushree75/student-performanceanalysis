# Task 1: Data Analysis with Python

## Objective
The objective of this task is to analyze student performance data using Python.
The dataset is explored, cleaned, analyzed, and visualized to understand
factors affecting student academic performance.


```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/student-mat.csv", sep=';')
# Display first 5 rows
df.head()
```

```python
# Dataset information
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

### Average Final Marks


```python
df['G3'].mean()
```

### Gender-wise Performance Analysis


```python
df.groupby('sex')['G3'].mean()
```

### Effect of Study Time on Final Grade


```python
df.groupby('studytime')['G3'].mean()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['G3'])
plt.xlabel("Final Marks (G3)")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades")
plt.show()
```

```python
import seaborn as sns

sns.barplot(x='sex', y='G3', data=df)
plt.title("Gender vs Final Grade")
plt.show()
```

```python
sns.boxplot(x='studytime', y='G3', data=df)
plt.title("Study Time vs Final Grade")
plt.show()
```

## Conclusion
- The dataset contains 395 student records with no missing values.
- The average final grade (G3) is approximately 10.4.
- Male students scored slightly higher than female students on average.
- Students with higher study time tend to achieve better final grades.
- Study time has a positive impact on academic performance.


```python
# Correlation between study time and final grade
df[['studytime', 'G3']].corr()
```

This shows a positive correlation between study time and final grades.
